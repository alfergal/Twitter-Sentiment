```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset sin preprocesar (asumiendo que ya lo has descargado)
df = pd.read_csv("../data/raw/training.1600000.processed.noemoticon.csv", encoding="latin-1", header=None)
df = df[[0, 5]]
df.columns = ["target", "text"]

# Mostrar las clases √∫nicas y sus conteos
conteo = df["target"].value_counts().sort_index()
print("Distribuci√≥n de clases:")
print(conteo)

# Gr√°fico de barras
plt.figure(figsize=(6, 4))
sns.barplot(x=conteo.index, y=conteo.values)
plt.title("Distribuci√≥n de etiquetas en Sentiment140")
plt.xlabel("Etiqueta (0 = negativo, 4 = positivo)")
plt.ylabel("N√∫mero de tweets")
plt.xticks([0, 1])
plt.tight_layout()
plt.show()
```
